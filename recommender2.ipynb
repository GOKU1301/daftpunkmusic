```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler




# Extract the relevant column for clustering
X = result[['views','likes','genre']]

# Standardize the features using StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Specify the number of clusters
n_clusters = 11

# Perform K-means clustering
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
result['cluster'] = kmeans.fit_predict(X_scaled)

# Create a dictionary with cluster number as key and list of cluster members as value
cluster_dict = {}
for cluster_num in range(n_clusters):
    members = result[result['cluster'] == cluster_num]['title'].tolist()
    cluster_dict[cluster_num] = members

# Print the DataFrame with cluster information
print("\nDataFrame with Cluster Information:")
print(result)

# Print the Cluster Dictionary
print("\nCluster Dictionary:")
print(cluster_dict)
```

```python
# Assuming you have a dictionary with song title IDs as keys and song titles as values
# song_title_dict = {
#     101: 'Song1',
#     102: 'Song2',
#     103: 'Song3',
#     104: 'Song4',
#     105: 'Song5',
#     106: 'Song6',
#     107: 'Song7',
#     108: 'Song8',
#     109: 'Song9',
# }

# Another dictionary with cluster IDs and lists of title IDs
# cluster_titles_dict = {
#     1: [101, 103, 107],
#     2: [102, 105, 108],
#     3: [104, 106, 109],
# }

# Function to recommend songs based on the number of songs from the same cluster in a playlist
def recommend_songs_in_playlist(playlist, song_title_dict, cluster_titles_dict, threshold=3):
    cluster_counts = {}

    # Count the number of songs from each cluster in the playlist
    for title_id in playlist:
        for cluster_id, title_ids in cluster_titles_dict.items():
            if title_id in title_ids:
                cluster_counts[cluster_id] = cluster_counts.get(cluster_id, 0) + 1

    # Recommend songs from clusters with more than the threshold number of songs
    recommended_titles = []
    for cluster_id, count in cluster_counts.items():
        if count > threshold and cluster_id in cluster_titles_dict:
            recommended_titles.extend(cluster_titles_dict[cluster_id])

    # Convert title IDs back to song titles
    recommended_songs = [song_title_dict[title_id] for title_id in recommended_titles]

    return recommended_songs

# Example playlist with title IDs
playlist = [1,7,8,9,1,4,6,8,3,2]

# Get recommended songs for the playlist
recommendations = recommend_songs_in_playlist(playlist,play_dict ,cluster_dict )

# Print the recommendations
print(f'Recommended Songs: {recommendations}')
```
